```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#wyznaczenie danych dla wykresu
t = np.arange(0.0,2.0,0.01)
s = 1 + np.sin(2*np.pi *t)
```

```python
fig,ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel='czas [s]', ylabel = 'napięcie prądu [mv]', title='pomiar napęcia prądu w czasie...')
ax.grid(True)
fig.savefig('test.png')
plt.show()
```
